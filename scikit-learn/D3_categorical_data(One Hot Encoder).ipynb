```python
import pandas as pd

d = {
    'sales': [100000, 222000, 1000000, 522000, 111111,222222,1111111,20000],
    'city': ['Tampa', 'Tampa', 'Orlando', 'Jacksonville', 'Miami','Jacksonville','Miami','Miami'],
    'size': ['Small', 'Medium', 'Large', 'Large', 'Small','Medium','Large','Small']
}

df = pd.DataFrame(data=d)

df.head(8)
```

```python
df['city'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder   
```

1. OneHotEncoder(...) Nedir?
Bilgisayarlar "Kırmızı", "Mavi", "Yeşil" gibi kelimelerden anlamaz; sayılarla konuşur. One-Hot Encoding, her kategoriyi ayrı bir sütuna dönüştürür.

Örnek: "Renk" sütununda 3 seçenek varsa, bunu 3 farklı sütuna (Renk_Kırmızı, Renk_Mavi, Renk_Yeşil) böler. O satır hangi renkse oraya 1, diğerlerine 0 yazar.


""" handle_unknown='ignore'	Hayat kurtarır. Eğer model eğitilirken görmediği yeni bir kategoriyle (mesela test verisinde "Sarı" gelirse) karşılaşırsa hata vermek yerine o satır için tüm sütunlara 0 basar ve geçer.


sparse_output=False	Normalde bu işlem "sparse matrix" (seyrek matris) döndürür (bellekten tasarruf için). Ama biz bunu False yaparak doğrudan okunabilir, yoğun bir dizi (numpy array) veya tablo istiyoruz.


.set_output(transform='pandas')	En can alıcı nokta burası. Eskiden bu işlemden sonra sütun isimleri kaybolur, elimizde isimsiz bir s
sayı yığını kalırdı. Bu komut sayesinde çıktı doğrudan bir Pandas DataFrame olarak döner ve sütun isimleri (örn: Renk_Kırmızı) otomatik eklenir."""


```python
ohe=OneHotEncoder(handle_unknown='ignore', sparse_output=False).set_output(transform='pandas')

#sparse_outpu dataframe olarak çıksın diye
```

```python
ohetransform= ohe.fit_transform(df[['city']])
ohetransform
```

```python
df.head()
```

```python
df=pd.concat([df,ohetransform], axis=1).drop(columns='city')
```
