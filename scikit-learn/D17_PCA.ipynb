```python
import pandas as pd 
import numpy as np
```

```python
df=pd.read_csv('../data/baseball.csv')
```

```python

pd.set_option('display.max_columns',None)
df.head()
```

```python
df.drop(columns=['Tm'], inplace=True)
```

```python
df.shape
```

```python
X=df.iloc[:,1:28]
y=df.iloc[:,28]
y[0]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.2, random_state=21)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train.columns
```

```python
X_train= scaler.fit_transform(X_train)
X_train=pd.DataFrame(X_train,columns=['BatAge', 'R/G', 'G', 'PA', 'AB', 'R', 'H', '2B', '3B', 'HR', 'RBI',
       'SB', 'CS', 'BB', 'SO', 'BA', 'OBP', 'SLG', 'OPS', 'OPS+', 'TB', 'GDP',
       'HBP', 'SH', 'SF', 'IBB', 'LOB'])
X_train.head()
```

```python
X_train.describe().round(3)
```

```python
from sklearn.decomposition import PCA
```

```python
pca1=PCA()
```

```python
X_pca1=pca1.fit_transform(X_train)
X_pca1.shape
```

```python
pca1.explained_variance_ratio_
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
plt.bar(range(1, len(pca1.explained_variance_)+1),pca1.explained_variance_)
plt.ylabel("Explained variance")
plt.xlabel("Components")

plt.plot(range(1,len(pca1.explained_variance_)+1),
         np.cumsum(pca1.explained_variance_),
         c='red',
         label='Cumulative Explained Variance'
         )
plt.legend(loc='upper left')
```

```python
pca2=PCA(0.95)
```

```python
X_pca2= pca2.fit_transform(X_train)
X_pca2.shape
```

```python
pca2.explained_variance_ratio_
```

```python
pca2c=PCA(n_components=2)
```

```python
X_pca2c=pca2.fit_transform(X_train)
```

```python

plt.figure()
scatter= plt.scatter(X_pca2c[:,0], X_pca2c[:,1], c=y_train, cmap='inferno')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.colorbar(scatter, label='Playoffs')
plt.show()
```

```python
pca3c=PCA(n_components=3)
```

```python
X_pca3c= pca3c.fit_transform(X_train)
```

```python
pca3c.explained_variance_ratio_
```

```python
from mpl_toolkits.mplot3d import Axes3D  # Import the 3D plotting toolkit

# Create a figure and a 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Create the 3D scatter plot
ax.scatter(X_pca3c[:,0], X_pca3c[:,1], X_pca3c[:,2], c=y_train, cmap='inferno')

# Set labels for the axes
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# Show the plot
plt.colorbar(scatter, label='Playoffs')
plt.show()
```
