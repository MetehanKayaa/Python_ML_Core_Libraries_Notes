```python
import seaborn as sns 
import pandas as pd

df=sns.load_dataset('healthexp')
df.head(10)
```

```python
df=pd.get_dummies(data=df)
df.head()
```

```python
X=df.drop(columns=['Life_Expectancy'])
y=df['Life_Expectancy']
```

```python
from sklearn.model_selection import train_test_split
X_Train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8,random_state=54)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor(random_state=34)
```

```python
rfr.fit(X_Train,y_train)
```

```python
y_pred=rfr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_squared_error(y_pred,y_test)
```

```python
r2_score(y_test,y_pred)
```

```python
import optuna
```

```python
from sklearn.model_selection import cross_val_score
```

```python
def objective(trial):
    n_estimators= trial.suggest_int('n_estimators',100,1000)
    max_depth=trial.suggest_int('max_depth',10,50)
    min_samples_split=trial.suggest_int('min_samples_split',2,32)
    min_samples_leaf=trial.suggest_int('min_samples_leaf',1,32)
    
    model=RandomForestRegressor(
        n_estimators=n_estimators,
        max_depth=max_depth,
        min_samples_split=min_samples_split,
        min_samples_leaf=min_samples_leaf,
    )
    
    score=cross_val_score(model,X_Train,y_train, cv=5,scoring='neg_mean_squared_error',n_jobs=-1).mean()
    return score 
```

```python
study=optuna.create_study(direction='maximize',sampler=optuna.samplers.RandomSampler(seed=42))
```

```python
study.optimize(objective,n_trials=200)
```

```python
study.best_params
```

```python
best_params=study.best_params
```

```python
best_n_estimator=best_params['n_estimators']
best_max_depth=best_params['max_depth']
best_min_samples_split=best_params['min_samples_split']
best_min_samples_leaf=best_params['min_samples_leaf']
```

```python
best_model=RandomForestRegressor(
    n_estimators=best_n_estimator,
                                 max_depth=best_max_depth,
                                 min_samples_split=best_min_samples_split,
                                 min_samples_leaf=best_min_samples_leaf)
```

```python
best_model.fit(X_Train,y_train)
```

```python
y_pred=best_model.predict(X_test)
```

```python
mean_absolute_error(y_pred,y_test)
```
