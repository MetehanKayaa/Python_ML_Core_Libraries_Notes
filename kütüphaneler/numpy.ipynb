```python
import numpy as np

# 2x2'lik bir matris oluşturalım
A = np.array([[1, 2], 
              [3, 4]])

# Matrisin tersini hesaplayalım
A_ters = np.linalg.inv(A)

print("Orijinal Matris:\n", A)
print("\nMatrisin Tersi:\n", A_ters)
```

```python
import numpy as np

# Katsayılar matrisi (A)
A = np.array([[3, 1], 
              [1, 2]])

# Sonuçlar matrisi (b)
b = np.array([9, 8])

# Denklemi çözelim (Ax = b)
x = np.linalg.solve(A, b)

print(f"x değeri: {x[0]}")
print(f"y değeri: {x[1]}")
```

```python
import numpy as np

# Kare bir matris tanımlayalım
A = np.array([[4, -2],
              [1,  1]])

# Özdeğer ve Özvektörleri hesaplayalım
ozdegerler, ozvektorler = np.linalg.eig(A)

print("Özdeğerler (Lambda):")
print(ozdegerler)

print("\nÖzvektörler (V):")
print(ozvektorler)
```
