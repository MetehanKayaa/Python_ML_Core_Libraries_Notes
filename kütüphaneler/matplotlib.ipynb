CUSTOMIZATİON FOR PLOTSS


```python
years= [2014+x for x in range(8)]
income=[55,56,62,61,72,72,73,75]

income_ticks=list(range(50,81,2))


plt.plot(years,income)
plt.yticks(income_ticks,[f"${x}K USD" for x in income_ticks])
plt.title('Income of salary (in USD)', fontsize=20, fontname='Serif')
plt.xlabel('Years')
plt.ylabel('Yearly Income in USD')
plt.show()
```

MULTİ PLOT İN A 1 TİME

```python
x1,y1 = np.random.random(100), np.random.random(100)

x2,y2= np.arange(100) , np.random.random(100)
# arange (0 100 arası sayı üretir 0 1 2 ...99 kdr)
plt.figure(1)
plt.scatter(x1,y1)

plt.figure(2)
plt.plot(x2,y2)

plt.show()
```

```python
x=np.arange(100)

fig, axs= plt.subplots(2,2)

axs[0,0].plot(x,np.sin(x))
axs[0,0].set_title('Sine Wawe')


axs[0,1].plot(x,np.cos(x))
axs[0,1].set_title('Cosine Wawe')

axs[1,0].plot(x,np.random.random(100))
axs[1,0].set_title('Random func')

axs[1,1].plot(x,np.log(x))
axs[1,1].set_title('Log Wawe')

fig.suptitle('Matplotlib Subplots', fontsize=20)
plt.show()
# how to export
plt.savefig('fourthplots.png')
```
