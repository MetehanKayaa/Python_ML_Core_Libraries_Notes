```python
def categorize_athlete(row):
    if (row['height_cm'] <175) & (row['weight_kg']<70):
        return 'LightWeight'
    if row['height_cm'] <185 and row['weight_kg']<85:
        return 'MiddleWeight'
    else:
        return 'HeavyWeight'

df['Category']=df.apply(categorize_athlete,axis=1)
```
