```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x= np.linspace(0,10,100)
y=np.sin(x) + np.random.normal(0, 0.2,100)
```

```python
plt.plot(x,y)
plt.title('Line plot')
plt.xlabel('X- axis')
plt.ylabel('Y- axis')
plt.show()
```

```python
plt.scatter(x,y)
plt.title('Scatter Plot')
plt.xlabel('X- axis')
plt.ylabel('Y- axis')
plt.show()
```

```python
categories = ['A','B','C','D','E']
values=[15,7,10,12,8]
plt.bar(categories,values)
plt.title('BAR PLOT')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
plt.hist(y, bins=20)
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
labels=['Category A', 'Category B', 'Category C']
sizes=[40,30,30]
explode=[0.1,0,0]
plt.pie(sizes, labels=labels, explode=explode, autopct='%.2f%%')
plt.title('Pie Chart')
plt.show()
```

```python
plt.fill_between(x,y, alpha=0.4)

plt.xlabel('X-axis')
plt.ylabel('Y- axis')
plt.show()
```

```python
theta= np.linspace(0, 2*np.pi, 100)
r=2+np.sin(5*theta)

plt.polar(theta,r)

plt.show()
```

```python
x, y = np.meshgrid(np.linspace(-3,3, 100), np.linspace(-3,3,100))
u= -y
v=x

plt.streamplot(x,y,u,v)

plt.xlabel('X axis')
plt.ylabel('Y- axis')
plt.show()
```

```python
x= np.linspace(0,10,100)
y=np.sin(x) + np.random.normal(0, 0.2,100)

errors= abs(np.random.normal(0,0.2,100))

plt.errorbar(x,y, yerr=errors, fmt='o', capsize=3, ecolor='r')
plt.show()
```
